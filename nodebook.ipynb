```python
import json
import os
import pandas
```

```python
exps = []
for file in os.listdir("./experiment"):
    with open(os.path.join("./experiment", file)) as f:
        match file.split("-"):
            case model, paramnum, adaptername, datasetname_json:
                model = "decapoda-research/llama-7b-hf"
            case org, model, paramnum, hf, adaptername, datasetname_json:
                model = org + "/" + model
            case _:
                raise ValueError()

        dataset = datasetname_json.split(".")[0]
        exps.append((json.loads(f.read()), model, adaptername, dataset))
        model = None

anss = []
for obj_ in exps:
    obj, model, adaptername, dataset = obj_
    all_ = len(obj)
    currnt = len([c for c in obj if c["flag"]])
    anss.append(
        {
            "acc": currnt / all_,
            "dataset": dataset,
            "adapter": adaptername,
            "model": model,
        }
    )


df = pandas.DataFrame(anss).sort_values(
    by=[
        "model",
        "dataset",
        "adapter",
    ],
    ascending=True,
)
df
```
