```python
import json
import os
import pandas
```

```python
exps = []
for file in os.listdir("./experiment"):
    with open(os.path.join("./experiment",file)) as f:
        model,_,adaptername,dataset = file.split('-')
        dataset = dataset.split('.')[0]
        exps.append((json.loads(f.read()),model,adaptername,dataset))
anss = []
for obj_ in exps:
    obj,model,adaptername,dataset = obj_
    all_ = len(obj)
    currnt = len([c for c in obj if c['flag']])
    anss.append({
        "acc":currnt/all_,
        "dataset":dataset,
        "adapter":adaptername,
    })


df = pandas.DataFrame(anss).sort_values(by="dataset" , ascending=True) 
df
```
